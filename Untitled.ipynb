```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
import pandas as pd
from sklearn.metrics import silhouette_score

# -----------------------------
# 1️⃣ Prepare your dataset
# -----------------------------
data = {
    "business_id": list(range(101, 121)),
    "monthly_revenue": [5000, 12000, 8000, 15000, 7000, 20000, 3000, 18000, 9000, 4000,
                        11000, 6000, 14000, 7500, 22000, 5000, 16000, 8500, 13000, 9500],
    "num_employees": [5, 12, 8, 15, 7, 20, 3, 18, 9, 4,
                      10, 6, 14, 7, 22, 5, 16, 8, 13, 9],
    "years_in_business": [1, 5, 3, 7, 2, 10, 1, 8, 4, 2,
                          6, 3, 7, 2, 12, 1, 9, 4, 5, 3],
    "customer_rating": [4.5, 4.0, 4.2, 3.8, 4.1, 4.7, 3.5, 4.6, 4.3, 3.9,
                        4.0, 3.7, 4.4, 4.1, 4.8, 3.6, 4.5, 4.2, 4.3, 4.0],
    "monthly_expenses": [3000, 7000, 4000, 9000, 3500, 12000, 2000, 10000, 5000, 2500,
                         6500, 3200, 8500, 4000, 15000, 3000, 9500, 4200, 8000, 5000],
    "industry_type": ["Retail", "Food", "Tech", "Food", "Retail", "Tech", "Retail", "Food", "Tech", "Retail",
                      "Food", "Retail", "Tech", "Retail", "Tech", "Food", "Food", "Tech", "Retail", "Food"]
}

df = pd.DataFrame(data)

# -----------------------------
# 2️⃣ Define columns
# -----------------------------
numeric_cols = ['business_id','monthly_revenue','num_employees','years_in_business','customer_rating','monthly_expenses']
categorical_cols = ['industry_type']

# -----------------------------
# 3️⃣ Create ColumnTransformer
# This will scale numeric features and encode categorical ones
# -----------------------------
trans = ColumnTransformer([
    ('scale', StandardScaler(), numeric_cols),
    ('encode', OneHotEncoder(sparse_output=False), categorical_cols)
])

# -----------------------------
# 4️⃣ Transform the data
# -----------------------------
X_transformed = trans.fit_transform(df)

# -----------------------------
# 5️⃣ Evaluate different numbers of clusters
# We'll test 2 → 6 clusters
# -----------------------------
cluster_range = range(2, 7)
inertia_list = []        # Sum of squared distances (for elbow method)
silhouette_list = []     # Silhouette scores (higher is better)

for k in cluster_range:   
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = kmeans.fit_predict(X_transformed)
    
    # Calculate metrics
    inertia = kmeans.inertia_
    sil_score = silhouette_score(X_transformed, labels)
    
    inertia_list.append(inertia)
    silhouette_list.append(sil_score)
    
    print(f"Clusters={k} → Inertia={inertia:.2f}, Silhouette Score={sil_score:.3f}")

# -----------------------------
# 6️⃣ Visualize results
# - Elbow Method: look for the "bend" in inertia
# - Silhouette: pick the highest score
# -----------------------------
plt.figure(figsize=(12,5))

# Elbow Method
plt.subplot(1,2,1)
plt.plot(cluster_range, inertia_list, marker='o', linestyle='--', color='blue')
plt.xlabel("Number of Clusters")
plt.ylabel("Inertia (Sum of Squared Distances)")
plt.title("Elbow Method: Inertia vs Number of Clusters")
plt.grid(True)

# Silhouette Score
plt.subplot(1,2,2)
plt.plot(cluster_range, silhouette_list, marker='o', linestyle='--', color='green')
plt.xlabel("Number of Clusters")
plt.ylabel("Silhouette Score")
plt.title("Silhouette Scores vs Number of Clusters")
plt.grid(True)

plt.tight_layout()
plt.show()
```
