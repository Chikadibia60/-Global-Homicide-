```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

# Step 1: Create dataset using dictionary (6 columns)
data = {
    "age": [22, 25, 30, 35, 40, 45, 50, 55],
    "salary": [20000, 25000, 30000, 35000, 60000, 65000, 70000, 75000],
    "spending_score": [40, 50, 60, 65, 30, 25, 20, 15],
    "experience": [1, 2, 5, 7, 10, 12, 15, 18],
    "family_size": [1, 2, 2, 3, 4, 4, 5, 5],
    "savings": [5000, 7000, 10000, 15000, 30000, 35000, 40000, 45000]
}

df = pd.DataFrame(data)

# Step 2: Scale data (VERY IMPORTANT)
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)

# Step 3: Apply K-Means
model = KMeans(n_clusters=2)
labels = model.fit_predict(scaled_data)

# Step 4: Add cluster result to dataset
df["cluster"] = labels

print(df)


# Step 5: Visualize (using 2 columns for plotting)
plt.scatter(df["age"], df["salary"], c=labels)
plt.xlabel("Age")
plt.ylabel("Salary")
plt.title("Customer Segmentation")
plt.show()

# Analyze clusters
cluster_summary = df.groupby("cluster").mean()

print(cluster_summary)

for i in range(2):
    print(f"\nCluster {i} Data:")
    print(df[df["cluster"] == i])
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# 1. Load the dataset
iris = load_iris()
X = iris.data  # Features only (no labels for unsupervised learning!)

# 2. Standardize the features (important for K-Means)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 3. Apply K-Means Clustering
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
clusters = kmeans.fit_predict(X_scaled)

# 4. Visualize using PCA (reduce to 2D for plotting)
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# 5. Plot the results
plt.figure(figsize=(12, 5))

# Plot 1: Discovered clusters
plt.subplot(1, 2, 1)
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=clusters, cmap='viridis', edgecolors='k')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], 
            c='red', marker='X', s=200, label='Centroids')
plt.title('K-Means Clustering (Unsupervised)')
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
plt.legend()

# Plot 2: Actual labels (for comparison)
plt.subplot(1, 2, 2)
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=iris.target, cmap='viridis', edgecolors='k')
plt.title('Actual Species Labels (Ground Truth)')
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')

plt.tight_layout()
plt.show()

# 6. Evaluate (optional - just to see how well it worked)
from sklearn.metrics import adjusted_rand_score
ari_score = adjusted_rand_score(iris.target, clusters)
print(f"Adjusted Rand Index: {ari_score:.3f}")
print(f"Inertia (Within-cluster sum of squares): {kmeans.inertia_:.2f}")
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Step 1: Create dataset (6 columns)
data = {
    "age": [22, 25, 30, 35, 40, 45, 50, 55, 28, 32],
    "salary": [20000, 25000, 30000, 35000, 60000, 65000, 70000, 75000, 28000, 32000],
    "spending_score": [40, 50, 60, 65, 30, 25, 20, 15, 55, 58],
    "experience": [1, 2, 5, 7, 10, 12, 15, 18, 3, 6],
    "family_size": [1, 2, 2, 3, 4, 4, 5, 5, 2, 3],
    "savings": [5000, 7000, 10000, 15000, 30000, 35000, 40000, 45000, 8000, 12000]
}

df = pd.DataFrame(data)

# Step 2: Scale the features
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)

# Step 3: Apply K-Means clustering
model = KMeans(n_clusters=2)
df['cluster'] = model.fit_predict(scaled_data)

# Step 4: Analyze clusters
print("Cluster Summary:\n", df.groupby('cluster').mean())

# Step 5: View data per cluster
for i in range(2):
    print(f"\nCluster {i} members:")
    print(df[df['cluster'] == i])

# Step 6: Visualize clusters (Age vs Salary)
plt.scatter(df['age'], df['salary'], c=df['cluster'], cmap='viridis')
plt.scatter(model.cluster_centers_[:, 0]*scaled_data.std(axis=0)[0] + scaled_data.mean(axis=0)[0], 
            model.cluster_centers_[:, 1]*scaled_data.std(axis=0)[1] + scaled_data.mean(axis=0)[1],
            marker='X', s=200, color='red')
plt.xlabel("Age")
plt.ylabel("Salary")
plt.title("Customer Segmentation")
plt.show()
```

```python
# Import libraries
import pandas as pd
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.cluster import KMeans, DBSCAN
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
import seaborn as sns

# -----------------------------
# 1. Create dictionary dataset
# -----------------------------
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Frank', 'Grace', 'Helen'],
    'Age': [25, 30, 35, 40, 22, 28, 32, 38],
    'Salary': [50000, 60000, 65000, 70000, 48000, 59000, 62000, 71000],
    'Department': ['HR','IT','IT','Finance','HR','IT','Finance','Finance']
}

df = pd.DataFrame(data)

# -----------------------------
# 2. Encode categorical data
# -----------------------------
encoder = OneHotEncoder(sparse_output=False)
dept_encoded = encoder.fit_transform(df[['Department']])
dept_df = pd.DataFrame(dept_encoded, columns=encoder.get_feature_names_out(['Department']))

# Combine numeric and encoded features
X = pd.concat([df[['Age','Salary']], dept_df], axis=1)

# -----------------------------
# 3. Scale numeric features
# -----------------------------
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# -----------------------------
# 4. Apply K-Means
# -----------------------------
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans_labels = kmeans.fit_predict(X_scaled)

# Silhouette Score for K-Means
kmeans_silhouette = silhouette_score(X_scaled, kmeans_labels)
print(f"K-Means Silhouette Score: {kmeans_silhouette:.3f}")

# -----------------------------
# 5. Apply DBSCAN
# -----------------------------
dbscan = DBSCAN(eps=1.5, min_samples=2)
dbscan_labels = dbscan.fit_predict(X_scaled)

# Silhouette Score for DBSCAN (only if >1 cluster)
if len(set(dbscan_labels)) > 1 and -1 not in set(dbscan_labels):
    dbscan_silhouette = silhouette_score(X_scaled, dbscan_labels)
else:
    dbscan_silhouette = "Cannot compute (1 cluster or noise)"
print(f"DBSCAN Silhouette Score: {dbscan_silhouette}")

# -----------------------------
# 6. Visualization
# -----------------------------
# Reduce to 2D using first two features for simple plot
plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
sns.scatterplot(x=X_scaled[:,0], y=X_scaled[:,1], hue=kmeans_labels, palette='Set1')
plt.title("K-Means Clusters")
plt.xlabel("Age (scaled)")
plt.ylabel("Salary (scaled)")

plt.subplot(1,2,2)
sns.scatterplot(x=X_scaled[:,0], y=X_scaled[:,1], hue=dbscan_labels, palette='Set2')
plt.title("DBSCAN Clusters")
plt.xlabel("Age (scaled)")
plt.ylabel("Salary (scaled)")

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.pipeline import Pipeline
import matplotlib.pyplot as plt

# 1️⃣ Dataset
data = {
    "age": [22, 25, 30, 35, 40, 45, 50, 55, 28, 32],
    "salary": [20000, 25000, 30000, 35000, 60000, 65000, 70000, 75000, 28000, 32000],
    "spending_score": [40, 50, 60, 65, 30, 25, 20, 15, 55, 58],
    "experience": [1, 2, 5, 7, 10, 12, 15, 18, 3, 6],
    "family_size": [1, 2, 2, 3, 4, 4, 5, 5, 2, 3],
    "savings": [5000, 7000, 10000, 15000, 30000, 35000, 40000, 45000, 8000, 12000]
}
df = pd.DataFrame(data)

# 2️⃣ Pipeline
pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('kmeans', KMeans(random_state=42))
])

# 3️⃣ Manual hyperparameter tuning (unsupervised)
best_inertia = float('inf')
best_k = None
for k in range(2, 6):  # try clusters from 2 to 5
    pipeline.set_params(kmeans__n_clusters=k)
    pipeline.fit(df)
    inertia = pipeline.named_steps['kmeans'].inertia_
    print(f"k={k}, inertia={inertia}")
    if inertia < best_inertia:
        best_inertia = inertia
        best_k = k

# 4️⃣ Fit best model
pipeline.set_params(kmeans__n_clusters=best_k)
pipeline.fit(df)
df['cluster'] = pipeline.named_steps['kmeans'].labels_

print(f"\nBest number of clusters: {best_k}")
print("\nCluster summary:\n", df.groupby('cluster').mean())

# 5️⃣ Visualize clusters (Age vs Salary)
plt.scatter(df['age'], df['salary'], c=df['cluster'], cmap='viridis')
plt.scatter(pipeline.named_steps['kmeans'].cluster_centers_[:, 0]*df[['age', 'salary']].std().iloc[0] + df['age'].mean(),
            pipeline.named_steps['kmeans'].cluster_centers_[:, 1]*df[['age', 'salary']].std().iloc[1] + df['salary'].mean(),
            marker='X', s=200, color='red')
plt.xlabel("Age")
plt.ylabel("Salary")
plt.title("Customer Segmentation")
plt.show()
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score
import matplotlib.pyplot as plt

# 1️⃣ Dataset
data = {
    "age": [22, 25, 30, 35, 40, 45, 50, 55, 28, 32],
    "salary": [20000, 25000, 30000, 35000, 60000, 65000, 70000, 75000, 28000, 32000],
    "spending_score": [40, 50, 60, 65, 30, 25, 20, 15, 55, 58],
    "experience": [1, 2, 5, 7, 10, 12, 15, 18, 3, 6],
    "family_size": [1, 2, 2, 3, 4, 4, 5, 5, 2, 3],
    "savings": [5000, 7000, 10000, 15000, 30000, 35000, 40000, 45000, 8000, 12000]
}
df = pd.DataFrame(data)

# 2️⃣ Transformer: scale all numerical columns
transformer = ColumnTransformer([
    ("num", StandardScaler(), df.columns)
])

# 3️⃣ Pipeline: transformer + KMeans
pipeline = Pipeline([
    ("transformer", transformer),
    ("kmeans", KMeans(n_clusters=2, random_state=42))  # choose n_clusters manually
])

# 4️⃣ Fit the model
pipeline.fit(df)

# 5️⃣ Predict clusters
df["cluster"] = pipeline.predict(df)

# 6️⃣ Calculate clustering metrics
labels = df["cluster"]
sil_score = silhouette_score(df, labels)
ch_score = calinski_harabasz_score(df, labels)
db_score = davies_bouldin_score(df, labels)

print("Clustered Data:\n", df)
print("\nMetrics:")
print(f"Silhouette Score: {sil_score:.3f} (higher is better)")
print(f"Calinski-Harabasz Index: {ch_score:.3f} (higher is better)")
print(f"Davies-Bouldin Index: {db_score:.3f} (lower is better)")

# 7️⃣ Visualize clusters (Age vs Salary)
plt.scatter(df["age"], df["salary"], c=df["cluster"], cmap="viridis")
# Transform cluster centers back to original scale for plotting
centers_scaled = pipeline.named_steps["kmeans"].cluster_centers_
plt.scatter(
    centers_scaled[:, 0] * df["age"].std() + df["age"].mean(),
    centers_scaled[:, 1] * df["salary"].std() + df["salary"].mean(),
    marker="X", s=200, color="red"
)
plt.xlabel("Age")
plt.ylabel("Salary")
plt.title("Customer Segmentation")
plt.show()
```
